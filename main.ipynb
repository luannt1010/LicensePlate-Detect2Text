```python
import warnings
import torch
from torch import optim
from torch.utils.data import DataLoader
from torchvision import transforms
import helper_utils
warnings.filterwarnings("ignore")
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)
```

```python
root_dir = "my_dataset"
batch_size = 2
val_factor, test_factor = 0.1, 0.1
transform = transforms.Compose([transforms.ToTensor()])
dataset, train_dataset, val_dataset, test_dataset = helper_utils.create_dataset_splits(root_dir, val_factor, test_factor, transform)
train_loader, val_loader, test_loader = helper_utils.create_dataloaders(train_dataset, val_dataset, test_dataset, batch_size)
print(f"Training set size:        {len(train_dataset)}")
print(f"Validation set size:      {len(val_dataset)}")
print(f"Test set size:            {len(test_dataset)}")
```

```python
class_map = {0: "background", 1: "car", 2: "motorcycle", 3: "license_plate"}
img, target = train_dataset[1]
print(target)
helper_utils.draw_bbox(img, target, class_map)
```

```python
num_classes = 4
save_path = "checkpoints/best_checkpoint_model.pth"
patience = 2
model = helper_utils.load_model(num_classes)
model = model.to(device)
EPOCHS = 6
params = [p for p in model.parameters() if p.requires_grad]
optimizer = optim.SGD(params=params, momentum=0.9, weight_decay=0.0005, lr=0.001)
model, metrics = helper_utils.training_loop(model, train_loader, val_loader, optimizer, EPOCHS, save_path, patience, device)
helper_utils.visualize_metrics(metrics)
```
